````python

import pandas as pd  # DataFrame récapitulatif de comparaison des modèles
model_df = pd.DataFrame({
    'Modèle': ['YOLO12m', 'YOLO12s', 'YOLO12n', 'YOLO11m', 'RT-DETR'],
    'Précision (P)': [0.652, 0.513, 0.456, 0.627, 0.575],
    'Rappel (R)': [0.538, 0.402, 0.349, 0.543, 0.418],
    'mAP50': [0.570, 0.411, 0.350, 0.555, 0.414],
    'mAP50-95': [0.366, 0.245, 0.202, 0.352, 0.236],
    'FPS': [39.53, 222.22, 285.71, 55.25, 40.65],
    'GFLOPs': [67.2, 21.2, 6.3, 67.7, 103.5]
})

model_df
````

````python
# Normaliser les données pour une meilleure visualisation comparative
# Créer un DataFrame pour le barplot groupé
import numpy as np

# Normalisation des métriques pour une échelle comparable (0-1)
model_df_norm = model_df.copy()
model_df_norm['mAP50_norm'] = model_df['mAP50']  # Déjà entre 0 et 1
model_df_norm['FPS_norm'] = model_df['FPS'] / model_df['FPS'].max()
model_df_norm['GFLOPs_norm'] = model_df['GFLOPs'] / model_df['GFLOPs'].max()

# Configuration du barplot groupé
x = np.arange(len(model_df['Modèle']))
width = 0.25

fig, ax = plt.subplots(figsize=(14, 7))

# Créer les trois groupes de barres
bars1 = ax.bar(x - width, model_df_norm['mAP50_norm'], width, label='mAP50', color='#3498db', alpha=0.8)
bars2 = ax.bar(x, model_df_norm['FPS_norm'], width, label='FPS (normalisé)', color='#2ecc71', alpha=0.8)
bars3 = ax.bar(x + width, model_df_norm['GFLOPs_norm'], width, label='GFLOPs (normalisé)', color='#e74c3c', alpha=0.8)

# Ajouter les valeurs réelles sur les barres
for i, (m50, fps, gflops) in enumerate(zip(model_df['mAP50'], model_df['FPS'], model_df['GFLOPs'])):
    ax.text(i - width, model_df_norm['mAP50_norm'].iloc[i] + 0.02, f'{m50:.3f}', 
            ha='center', va='bottom', fontsize=9, fontweight='bold')
    ax.text(i, model_df_norm['FPS_norm'].iloc[i] + 0.02, f'{fps:.1f}', 
            ha='center', va='bottom', fontsize=9, fontweight='bold')
    ax.text(i + width, model_df_norm['GFLOPs_norm'].iloc[i] + 0.02, f'{gflops:.1f}', 
            ha='center', va='bottom', fontsize=9, fontweight='bold')

# Personnalisation
ax.set_xlabel('Modèle', fontsize=13, fontweight='bold')
ax.set_ylabel('Valeur Normalisée (0-1)', fontsize=13, fontweight='bold')
ax.set_title('Comparaison des Modèles - mAP50, FPS et GFLOPs', fontsize=15, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(model_df['Modèle'], rotation=45, ha='right')
ax.legend(fontsize=11, loc='upper right')
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_ylim(0, 1.15)

plt.tight_layout()
plt.show()
````
