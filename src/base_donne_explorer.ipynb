```python
# Croisement des colonnes entre les deux datasets
import pandas as pd

print("=== ANALYSE DE CROISEMENT DES COLONNES ===")
print()

# Charger les deux fichiers de colonnes
file1_path = "../data/base_donne_cols.csv"
file2_path = "../doc/insd_columns.csv"

try:
    # Lire les colonnes du premier fichier (BASE_DONNEES)
    print(f"📁 Lecture du fichier 1: {file1_path}")
    df_base_donnees = pd.read_csv(file1_path, header=None, names=['colonne'])
    colonnes_base_donnees = set(df_base_donnees['colonne'].astype(str).str.strip())
    print(f"   ✅ {len(colonnes_base_donnees)} colonnes trouvées dans BASE_DONNEES")
    
    # Lire les colonnes du deuxième fichier (INSD)
    print(f"📁 Lecture du fichier 2: {file2_path}")
    df_insd = pd.read_csv(file2_path, header=None, names=['colonne'])
    colonnes_insd = set(df_insd['colonne'].astype(str).str.strip())
    print(f"   ✅ {len(colonnes_insd)} colonnes trouvées dans INSD")
    
    print()
    print("="*80)
    print("ANALYSE COMPARATIVE")
    print("="*80)
    
    # 1. Colonnes communes
    colonnes_communes = colonnes_base_donnees.intersection(colonnes_insd)
    print(f"🤝 COLONNES COMMUNES ({len(colonnes_communes)} colonnes):")
    if colonnes_communes:
        for i, col in enumerate(sorted(colonnes_communes), 1):
            print(f"   {i:3d}. {col}")
    else:
        print("   Aucune colonne commune trouvée")
    
    print()
    
    # 2. Colonnes uniquement dans BASE_DONNEES
    colonnes_uniquement_base = colonnes_base_donnees - colonnes_insd
    print(f"📊 COLONNES UNIQUEMENT dans BASE_DONNEES ({len(colonnes_uniquement_base)} colonnes):")
    if colonnes_uniquement_base:
        for i, col in enumerate(sorted(colonnes_uniquement_base), 1):
            print(f"   {i:3d}. {col}")
    else:
        print("   Aucune colonne unique à BASE_DONNEES")
    
    print()
    
    # 3. Colonnes uniquement dans INSD
    colonnes_uniquement_insd = colonnes_insd - colonnes_base_donnees
    print(f"📈 COLONNES UNIQUEMENT dans INSD ({len(colonnes_uniquement_insd)} colonnes):")
    if colonnes_uniquement_insd:
        for i, col in enumerate(sorted(colonnes_uniquement_insd), 1):
            print(f"   {i:3d}. {col}")
    else:
        print("   Aucune colonne unique à INSD")
    
    print()
    print("="*80)
    print("STATISTIQUES RÉSUMÉES")
    print("="*80)
    
    total_colonnes_uniques = len(colonnes_base_donnees.union(colonnes_insd))
    taux_overlap = (len(colonnes_communes) / total_colonnes_uniques * 100) if total_colonnes_uniques > 0 else 0
    
    print(f"📊 Total colonnes BASE_DONNEES: {len(colonnes_base_donnees)}")
    print(f"📈 Total colonnes INSD: {len(colonnes_insd)}")
    print(f"🤝 Colonnes communes: {len(colonnes_communes)}")
    print(f"📊 Colonnes uniques BASE_DONNEES: {len(colonnes_uniquement_base)}")
    print(f"📈 Colonnes uniques INSD: {len(colonnes_uniquement_insd)}")
    print(f"🔢 Total colonnes uniques (union): {total_colonnes_uniques}")
    print(f"📈 Taux de recouvrement: {taux_overlap:.2f}%")
    
    # Pourcentages par rapport à chaque dataset
    print()
    print("📊 POURCENTAGES DE RECOUVREMENT:")
    if len(colonnes_base_donnees) > 0:
        pct_communes_base = len(colonnes_communes) / len(colonnes_base_donnees) * 100
        print(f"   Colonnes communes / BASE_DONNEES: {pct_communes_base:.2f}%")
    
    if len(colonnes_insd) > 0:
        pct_communes_insd = len(colonnes_communes) / len(colonnes_insd) * 100
        print(f"   Colonnes communes / INSD: {pct_communes_insd:.2f}%")
    
    print()
    print("="*80)
    print("SAUVEGARDE DES RÉSULTATS")
    print("="*80)
    
    # Créer un DataFrame de résultats détaillé
    resultats = []
    
    # Ajouter les colonnes communes
    for col in sorted(colonnes_communes):
        resultats.append({
            'colonne': col,
            'dans_base_donnees': 'OUI',
            'dans_insd': 'OUI',
            'statut': 'COMMUNE'
        })
    
    # Ajouter les colonnes uniquement dans BASE_DONNEES
    for col in sorted(colonnes_uniquement_base):
        resultats.append({
            'colonne': col,
            'dans_base_donnees': 'OUI',
            'dans_insd': 'NON',
            'statut': 'UNIQUE_BASE_DONNEES'
        })
    
    # Ajouter les colonnes uniquement dans INSD
    for col in sorted(colonnes_uniquement_insd):
        resultats.append({
            'colonne': col,
            'dans_base_donnees': 'NON',
            'dans_insd': 'OUI',
            'statut': 'UNIQUE_INSD'
        })
    
    # Créer le DataFrame et sauvegarder
    df_resultats = pd.DataFrame(resultats)
    
    # Sauvegarder le résultat détaillé
    output_file = "../data/comparaison_colonnes.csv"
    df_resultats.to_csv(output_file, index=False)
    print(f"💾 Résultats détaillés sauvegardés dans: {output_file}")
    
    # Sauvegarder les listes séparées
    if colonnes_communes:
        pd.DataFrame(sorted(colonnes_communes), columns=['colonne']).to_csv(
            "../data/colonnes_communes.csv", index=False)
        print(f"🤝 Colonnes communes sauvegardées dans: ../data/colonnes_communes.csv")
    
    if colonnes_uniquement_base:
        pd.DataFrame(sorted(colonnes_uniquement_base), columns=['colonne']).to_csv(
            "../data/colonnes_unique_base_donnees.csv", index=False)
        print(f"📊 Colonnes uniques BASE_DONNEES sauvegardées dans: ../data/colonnes_unique_base_donnees.csv")
    
    if colonnes_uniquement_insd:
        pd.DataFrame(sorted(colonnes_uniquement_insd), columns=['colonne']).to_csv(
            "../data/colonnes_unique_insd.csv", index=False)
        print(f"📈 Colonnes uniques INSD sauvegardées dans: ../data/colonnes_unique_insd.csv")
    
    print()
    print("="*80)
    print("APERÇU DU TABLEAU DE COMPARAISON")
    print("="*80)
    
    # Afficher un aperçu du tableau de résultats
    print("📋 Aperçu des résultats (premières et dernières lignes):")
    print()
    print("Premières 10 lignes:")
    print(df_resultats.head(10).to_string(index=False))
    
    if len(df_resultats) > 20:
        print("\n...")
        print("Dernières 10 lignes:")
        print(df_resultats.tail(10).to_string(index=False))
    
    # Statistiques par statut
    print()
    print("📊 RÉPARTITION PAR STATUT:")
    +6
    
    
    
    
    
    statut_counts = df_resultats['statut'].value_counts()
    for statut, count in statut_counts.items():
        print(f"   {statut}: {count} colonnes")
    
    print()
    print("✅ Analyse de croisement terminée avec succès!")
    
except FileNotFoundError as e:
    print(f"❌ Erreur: Fichier non trouvé - {e}")
    print("   Vérifiez que les fichiers existent aux chemins spécifiés")
    
except Exception as e:
    print(f"❌ Erreur lors de l'analyse: {e}")
    import traceback
    traceback.print_exc()
```

```python
inds_cols  =pd.read_csv("../data/insd/insd_columns.csv", header=None, names=['colonne'])
inds_cols = set(inds_cols['colonne'].astype(str).str.strip())
cols = [
    "Num_IFU_Contribuable.x",
    "ANNEE",
    "ETAT",
    "CODE_REG_FISC",
    "NOM_MINEFID",
    "DATE_IMMAT",
    "ANNEE_FISCAL"
]

print("=== VÉRIFICATION DE LA PRÉSENCE DES COLONNES DANS INSD ===")
print()
print(f"📊 Total colonnes INSD: {len(inds_cols)}")
print(f"🔍 Colonnes à vérifier: {len(cols)}")
print()

# Vérifier chaque colonne
resultats_verification = []
colonnes_trouvees = 0
colonnes_manquantes = 0

for i, colonne in enumerate(cols, 1):
    if colonne in inds_cols:
        status = "✅ TROUVÉE"
        colonnes_trouvees += 1
        dans_insd = True
    else:
        status = "❌ MANQUANTE"
        colonnes_manquantes += 1
        dans_insd = False
    
    print(f"{i:2d}. {colonne:<25} {status}")
    
    # Stocker le résultat
    resultats_verification.append({
        'colonne': colonne,
        'dans_insd': dans_insd,
        'statut': 'TROUVÉE' if dans_insd else 'MANQUANTE'
    })

print()
print("="*60)
print("RÉSUMÉ DE LA VÉRIFICATION")
print("="*60)
print(f"✅ Colonnes trouvées: {colonnes_trouvees}/{len(cols)} ({colonnes_trouvees/len(cols)*100:.1f}%)")
print(f"❌ Colonnes manquantes: {colonnes_manquantes}/{len(cols)} ({colonnes_manquantes/len(cols)*100:.1f}%)")

if colonnes_manquantes > 0:
    print()
    print("⚠️  COLONNES MANQUANTES:")
    for result in resultats_verification:
        if not result['dans_insd']:
            print(f"   ❌ {result['colonne']}")

print()
print("="*60)
print("RECHERCHE DE COLONNES SIMILAIRES")
print("="*60)

# Rechercher des colonnes similaires pour les manquantes
for result in resultats_verification:
    if not result['dans_insd']:
        colonne_recherchee = result['colonne'].lower()
        colonnes_similaires = []
        
        # Recherche par correspondance partielle
        for col_insd in inds_cols:
            col_insd_lower = col_insd.lower()
            
            # Vérifier si des mots-clés sont contenus
            mots_cles = colonne_recherchee.replace('_', ' ').replace('.', ' ').split()
            correspondances = sum(1 for mot in mots_cles if mot in col_insd_lower)
            
            if correspondances > 0:
                colonnes_similaires.append((col_insd, correspondances))
        
        if colonnes_similaires:
            # Trier par nombre de correspondances
            colonnes_similaires.sort(key=lambda x: x[1], reverse=True)
            print(f"\n🔍 Colonnes similaires à '{result['colonne']}':")
            for col_sim, score in colonnes_similaires[:5]:  # Top 5
                print(f"   📋 {col_sim} (score: {score})")

# Créer un DataFrame des résultats
df_verification = pd.DataFrame(resultats_verification)

# Sauvegarder les résultats
output_file = "../data/verification_colonnes_cibles.csv"
#df_verification.to_csv(output_file, index=False)
print()
print(f"💾 Résultats de vérification sauvegardés dans: {output_file}")

print()
print("✅ Vérification terminée!")
```
